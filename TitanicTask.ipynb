```python
train_dataFrame = train_dataFrame.drop(["Cabin"], axis=1)
```

```python
train_dataFrame['Age'].fillna(train_dataFrame['Age'].median(), inplace=True)
train_dataFrame['Embarked'].fillna(train_dataFrame['Embarked'].mode(), inplace=True)
```

```python
train_dataFrame = train_dataFrame.drop(["PassengerId", "Fare", "Ticket", "Name"], axis = 1)
```

```python
c_column = train_dataFrame.drop(train_dataFrame.select_dtypes(exclude=['object']), axis=1).columns
print(c_column)

enc1 = LabelEncoder()
train_dataFrame[c_column[0]] = enc1.fit_transform(train_dataFrame[c_column[0]].astype('str'))

enc2 = LabelEncoder()
train_dataFrame[c_column[1]] = enc2.fit_transform(train_dataFrame[c_column[1]].astype('str'))
```
